```python
import pandas as pd
import numpy as np
from numpy import random
ser1=pd.Series(data=random.randint(10,45,size=5),
index=['a','b','c','d','e'])
print("Series is")
print(ser1)
df=pd.DataFrame(data=np.arange(101,126).reshape(5,5),
index=['A','B','C','D','E'],
columns=['U','V','W','X','Y'])
print("Dataframeis")
print(df)
print("Columnwise accessing")
print(df['W']['A'])
print(df['W'])
print(df[['W','X','U']])
print("Rowwise accsessing")
print(df.loc['A']['X'])
print(df.loc['B'])
print(df.loc[['B','A']])
print(df.iloc[2]['X'])
print(df.iloc[1])
print(df.iloc[2:4])
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,6)
y = x**2
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title("SimplePlot")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(y,x)
plt.show()
```

```python
x=[1,2,3,4]
y=[i**2 for i in x]
plt.plot(x,y,label='x-square')
y=[i**3 for i in x]
plt.plot(x,y,label='x-cube')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('square-cube plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,6,8,10]
y=[i**2 for i in x]
plt.plot(x,y,marker='o',markerfacecolor='red')
plt.xlabel('Number')
plt.ylabel('Square of the number')
plt.title('Square Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,4,6,8,10]
y=[4,16,36,64,100]
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['A','B','C','D']
y=[50,25,100,75]
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[21,43,56,58,45,54,36,20,51,90,85]
plt.hist(x,bins=[0,25,50,75,100])
plt.xticks([0,25,50,75,100])
plt.show()
```
